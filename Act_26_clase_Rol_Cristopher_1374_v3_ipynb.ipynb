```python
# clase Rol, atributos: nombre, permisos, descripcion
# funciones: leerdatos(), mostrardatos()
# Crear 2 objetos: rol1, rol2

class Rol:
  def __init__(self, nombre, permisos, descripcion):
    self.nombre = nombre
    self.permisos = permisos
    self.descripcion = descripcion

  def leerdatos(self):
    print("Por favor, ingrese los datos del rol:")
    self.nombre = input("Ingrese el nombre del rol: ")
    self.permisos = input("Ingrese los permisos del rol: ")
    self.descripcion = input("Ingrese la descripción del rol: ")

  def mostrardatos(self):
    print("Datos del Rol:")
    print(f"Nombre: {self.nombre}")
    print(f"Permisos: {self.permisos}")
    print(f"Descripción: {self.descripcion}")

# Fin de la clase Rol

# Creando los objetos rol1, rol2

# Inicializamos los objetos
rol1 = Rol("", "", "") # Initialize with empty strings or default values
rol2 = Rol("", "", "")

print("Rol Numero 1")

rol1.leerdatos() # Call leerdatos on the instance
rol1.mostrardatos() # Call mostrardatos on the instance
print("-------------------------------------------------")

print("Rol Numero 2")
rol2.leerdatos()
rol2.mostrardatos()
print("-------------------------------------------------")

print("programa realizado por Cristopher Lopez NL = 33")
```
